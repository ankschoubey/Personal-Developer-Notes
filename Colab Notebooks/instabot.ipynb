```
#https://stackoverflow.com/questions/7698231/python-pil-draw-multiline-text-on-image

from PIL import Image, ImageDraw, ImageFont
import textwrap

def draw_multiple_line_text(image, text, font, text_color, text_start_height):
    '''
    From unutbu on [python PIL draw multiline text on image](https://stackoverflow.com/a/7698300/395857)
    '''
    draw = ImageDraw.Draw(image)
    image_width, image_height = image.size
    y_text = text_start_height
    lines = textwrap.wrap(text, width=40)
    for line in lines:
        line_width, line_height = font.getsize(line)
        draw.text(((image_width - line_width) / 2, y_text), 
                  line, font=font, fill=text_color)
        y_text += line_height

def make_image(text, filename):
    '''
    Testing draw_multiple_line_text
    '''
    #image_width
    image = Image.new('RGB', (1024, 1024), color = (0, 0, 0))
    fontsize = 55  # starting font size
    font = ImageFont.truetype("arial.ttf", fontsize)
    text_color = (200, 200, 200)
    text_start_height = 200
    draw_multiple_line_text(image, "Today in History", font, text_color, text_start_height)
    draw_multiple_line_text(image, text, font, text_color, 400)
    image.save(filename)
```
