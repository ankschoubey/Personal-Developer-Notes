```
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
```

```

class MyNN(torch.nn.Module):
    def __init__(self):
        super(self.__class__, self).__init__()
        self.lin1 = nn.Linear(100,10)
        self.lin2 = nn.Linear(10, 1)
    
    def forward(self, x):
        x = self.lin1(x)
        x = F.relu(x)

        x = self.lin2(x)
        x = F.sigmoid(x)

        return x
```

```
x = torch.randn(10,100)
n = MyNN()
```

```
x.shape
```

```
out = n(x)
out
```

```
target = torch.randn(10,1)
target
```

```
loss_func = nn.MSELoss()
```

```
loss = loss_func(out, target)
```

```
loss
```

```
n.zero_grad()
loss.backward()
```

```
lr = 0.01
for i in n.parameters():
    i.data.sub_(i.grad.data*lr)
```

```
optim = optim.SGD(n.parameters(), lr)
```

```
optim.zero_grad()
out = n(x)
loss = loss_func(out, target)
loss.backward()
optim.step()
```
