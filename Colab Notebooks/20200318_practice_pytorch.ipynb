```
import torch
```

```
import torch.nn as nn
```

```
class MyNN(nn.Module):
    def __init__(self):
        super(MyNN, self).__init__()
    
    def forward(self, x):
        return x
```

```
n = MyNN()
```

```
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
```

```
if torch.cuda.device_count() > 1:
    print('Using ', torch.cuda.device_count(), 'gpus')
    n = nn.DataParallel(n)
n.to(device)
```
