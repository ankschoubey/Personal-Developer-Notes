## Code

```
import pandas as pd
import pandas_profiling as pp
```

```
df = pd.read_csv('train.csv')
df.head()
```

```
report = pp.ProfileReport(df)
```

```
eda = report.description_set
eda.keys()
```

```
variable_info = eda['variables']
variable_info.head()
```

## Get Categorical Variables

```
cat_vars = variable_info[variable_info['type'] == 'CAT'].index
cat_vars
```

To be able to use it with fastai tabular just convert it into list

```
cat_vars = list(cat_vars)
cat_vars[:7]
```

## Get Continous Variables

```
cont_vars = variable_info[variable_info['type'] == 'NUM'].index
cont_vars
```

```
cont_vars = list(cont_vars)
cont_vars[:7]
```
