```
import torch
import matplotlib.pyplot as plt
```

# Variance

```
x = torch.randn(10)
x
```

```
plt.scatter(range(x.numel()),x)
```

```
mean = x.mean()
mean
```

```
x - mean
```

```
wrong_variance = (x-mean).mean()
wrong_variance, wrong_variance.round()
```

The reason that this is 0 is because positive and negatives cancel out. Therefore we need to make sure everything is positive

```
# mean absolute deviation

mad = (x-mean).abs().mean()
mad
```

```
(x-mean).abs()
```

```
# standard deviation

sd = (x-mean).pow(2).mean().sqrt()
sd
```
