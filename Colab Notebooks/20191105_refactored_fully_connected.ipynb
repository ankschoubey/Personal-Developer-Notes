```
import torch
from torch import nn
```

```
data = torch.rand(500,10)
mean, std = data.mean(), data.std()
mean, std
```

```
def normalize(data,train_mean, train_std): return (data-train_mean)/train_std
data = normalize(data, mean, std)
data.mean(), data.std()
```

```
lin1 = nn.Linear(10,1)
```

```
lin1.weight, lin1.weight.shape
```

```
lin1.bias
```

```
from torch.nn import init
```

```
init.kaiming_normal_(lin1.weight, mode='fan_in',nonlinearity='relu')
init.zeros_(lin1.bias)
```

```
lin1.weight.mean(), lin1.weight.std()
```

```
import torch.nn.functional as F
```

```
output = F.relu(lin1(data))
```

```
output.mean(), output.std()
```

```
lin1 = nn.Linear(10,1)
init.kaiming_uniform_(lin1.weight, mode='fan_out',nonlinearity='relu')
init.zeros_(lin1.bias)
```

```

output = F.relu(lin1(data))
output.mean(), output.std()
```
