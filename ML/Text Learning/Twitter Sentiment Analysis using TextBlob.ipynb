## Sentiment Analysis

Tags: What type of word it is. (Noun, Verb, Proper Noun, etc)

**Polarity**: 0 indicates neutral, +1 a very positive attitude and -1 a very negative attitude

**Subjectivity**: [0.0, 1.0] where 0.0 is very objective and 1.0 is very subjective.

```python
sentence = "Ankush is angry, he never gets any matches on tinder."
engine = TextBlob(sentence)

print('Tags',engine.tags)
```

# Twitter Sentiment Analysis

Get data for below values from apps.twitter.com and save it into string

```python
consumer_key = ''
consumer_secret = ''

access_token = ''
access_secret = ''
```

```python
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
```

```python
auth.set_access_token(access_token, access_secret)
```

```python
api = tweepy.API(auth)
```

```python
tweets = api.search('Machine Learning')
```

```python
for tweet in tweets[:2]:
    blob = TextBlob(tweet.text)
    print(tweet.text, blob.tags, blob.sentiment,sep='\n\n')
    print('***\n\n')
```
